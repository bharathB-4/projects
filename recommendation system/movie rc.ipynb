```python
c=fm["vote_average"].mean()
c
```

```python
m=fm["vote_count"].quantile(0.8)
m
```

```python
rc=fm[(fm["runtime"]>140) & (fm["revenue"]>310000000) & (fm["homepage"]).notna()]
rc.head()
```

```python
rc.shape
```

```python
rc["vote_count"]>=m
```

```python
rc=rc[rc["vote_count"]>=m]
rc.shape
```

```python
#imdb rating calculation
def wr(x,m=m,c=c):
    v=x["vote_count"]
    r=x["vote_average"]
    return (v/(v+m)*r)+(m/(m/(m+v)*c))
```

```python
#gives score
rc["score"]=rc.apply(wr,axis=1)
rc.head()
```

```python
rc=rc.sort_values("score",ascending=False)
```

```python
#sorting based on score
rc[["title","vote_count","vote_average","score"]].head()
```

```python
#creating visualization
d=fm.sort_values("vote_count",ascending=False)
plt.figure(figsize=(8,16))
plt.barh(d["title"].head(15),d["vote_count"].head(15),align='center',color='red')
plt.gca().invert_yaxis()
plt.xlabel("votes")
plt.title("must voted movies")
```
